```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
df_st = pd.read_csv("StudentsPerformance.csv")

# Display the first few rows
df_st.head()

# Drop missing values
df_st.dropna(inplace=True)
```

```python
# Function to create a bar plot for race/ethnicity
def create_race_ethnicity_plot(data):
    """
    Create a bar plot to visualize the distribution of students based on race/ethnicity.
    """
    plt.figure(figsize=(8, 6))
    data.groupby('race/ethnicity').size().plot(kind='barh', color=sns.color_palette('Dark2'))
    plt.xlabel('Number of Students')
    plt.ylabel('Race/Ethnicity')
    plt.title('Distribution of Students by Race/Ethnicity')
    plt.show()
```

```python
create_race_ethnicity_plot(df_st)
```

```python
# Function to create a scatter plot for math score vs reading score
def create_math_reading_plot(data):
    """
    Create a scatter plot between math scores and reading scores.

    Args:
    data (DataFrame): Input Data 'math score' and 'reading score' columns.
    """
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=data, x='math score', y='reading score', s=32, alpha=.8)
    plt.xlabel('Math Score')
    plt.ylabel('Reading Score')
    plt.title('Math Score vs Reading Score')
    plt.show()
```

```python
create_math_reading_plot(df_st)
```

```python
# Function to create a violin plot for gender vs math score
def create_gender_math_plot(data):
    """
    Create a violin plot of distribution of math scores across genders.

    Args:
    data (DataFrame): Input DataFrame containing 'gender' and 'math score' columns.
    """
    figsize = (8, 3 * len(data['gender'].unique()))
    plt.figure(figsize=figsize)
    sns.violinplot(data=data, x='math score', y='gender', inner='box', palette='Dark2')
    plt.xlabel('Math Score')
    plt.ylabel('Gender')
    plt.title('Distribution of Math Scores by Gender')
    plt.show()
```

```python

create_gender_math_plot(df_st)
```
